```python
import pandas as pd
import numpy as np

def read_in_csv(csv_file):
    # readin csv without headers so that the actual header can be processed later ( to add fillers)
    csv_data = pd.read_csv(csv_file, header=None, skip_blank_lines=True)

    # drop empty lines
    csv_data.dropna(axis = 0, how = 'all', inplace=True)

    # reset index
    csv_data.reset_index(inplace=True, drop=True)

    # forward-fill index 
    csv_data.fillna(method='ffill',inplace=True)

    # forward-fill columns. transpose twice as column-wise filling has not been found / implemented
    csv_data = csv_data.transpose()
    csv_data.fillna(method='ffill', inplace=True)
    csv_data = csv_data.transpose()

    # extract column, index & data to create a new data with the correct column & index settings
    header = csv_data.loc[0:2, 2:].values
    index = csv_data.loc[3:, 0:1].values.transpose()
    data = csv_data.loc[3:, 2:].astype(np.int).values  
    csv_data = pd.DataFrame(data, index=index.tolist(), columns=header.tolist())

    return csv_data

csv_data = read_in_csv('./CollegeGrades1 a.csv')
csv_data
```

```python

def fillna(row_list):
    '''Fill empty element with the first of its precedent non-empty values'''
    filler = ''
    for i,v in enumerate(row_list):
        if v == '':
            row_list[i] = filler
        else:
            filler = v
    return row_list
            
data = []
header = []
index = []
with open('./CollegeGrades1.csv', 'r') as f:
    i = 0
    line = f.readline().rstrip('\n')
    while line:
        row = line.split(',')
        if len(set(row)) > 1: 
            # To skip empty rows
            if i <= 2:
                row = fillna(row)
                header.append(row[2:])
            else:
                row_int = [int(e) for e in row[2:]]
                data.append(row[2:])
                # to fill row with the value of the previous row
                if row[0] == '' and i > 2:
                    row[0] = index[len(index)-1][0]
                if row[1] == '' and i > 2:
                    row[1] = index[len(index)-1][1]
                
                index.append(row[:2])
            i += 1
        line = f.readline().rstrip('\n')


# get rows that are forming hierarchical columns

# get index (transposed to meet the format for Dataframe)
index_trans = pd.DataFrame(index).transpose().values.tolist()

csv_data = pd.DataFrame(data, columns=header, index=index_trans)
csv_data.head()
```

## Merge 2 datasets
### Steps
1. Name Columns ['College', 'Subject', 'Status'] & Index ['Year', 'Gender'] for both DataFrame tables
2. Manipulate the second Dataset <br>
     a. Sort by 'Year' 
     b. Swap the 2 'Year' and 'Gender' indexies
     c. Replace index '16-17' to 'Year 1' and '17-18' to 'Year 2'
3. Join the 2 DataFrame tables together with pd.join() 

```python
# Assign column names to the hierarchical columns
csv_data_1.columns.names = ['College', 'Subject', 'Status']
csv_data_2.columns.names = ['College', 'Subject', 'Status']

# Name index with 'Gender' and 'Year'
csv_data_1.index.names=(['Year','Gender'])
csv_data_1
```

```python
# Name index with 'Gender' and 'Year'
csv_data_2.index.names=(['Gender','Year'])
csv_data_2
```

```python
# Sort the index by 'Year'
csv_data_2_sorted = csv_data_2.sort_index(level='Year')
csv_data_2_sorted

# swap Gender & Year columns
csv_data_2_sorted_swapped = csv_data_2_sorted.swaplevel('Gender','Year')

# change '16-17' to 'Year 1', '17-18' to 'Year 2'
mi_values = [['Year 1',e[1]] if e[0] == '16-17' else ['Year 2', e[1]] for e in csv_data_2_sorted_swapped.index.values]
mi_names = list(csv_data_2_sorted_swapped.index.names)
mi_values = np.array(mi_values).transpose()
multi_index = pd.MultiIndex.from_arrays(mi_values, names=mi_names)
csv_data_2_sorted_swapped.index = multi_index
multi_index
```

```python
csv_data_2_sorted_swapped
```

```python
# Join 2 Dataframe tables together
#pd.concat([csv_data_1, csv_data_2_sorted_swapped], axis=1, join='inner')
csv_data_joined = csv_data_1.join(csv_data_2_sorted_swapped)
csv_data_joined
```

```python
female = csv_data_joined_swapped.loc['F', :].unstack()
female_percent = female.sum(level=['College', 'Status'])/female.sum(level=['College'])
female_percent
```

```python
# The percentage of Female > Male
female_larger_percent=female_percent[female_percent > male_percent]

# The subjects where Female has better Pass rate than Male 
female_larger_percent[female_larger_percent.index.get_level_values(1) == 'P']
```
